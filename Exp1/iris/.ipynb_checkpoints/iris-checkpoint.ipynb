```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

columns_names=['sl','sw','pl','pw','class']
df=pd.read_csv('iris.csv',header=None,names=columns_names)
df.head()
```

```python

print("First 5 rows of the dataset:")
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sl'], df['sw']) 
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['pl'],df['pw'])
plt.title('Scatter Plot: Petal Length vs Petal width')
plt.xlabel('Petal Length')
plt.ylabel('Petal width')
plt.show
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sl'],df['pl'])
plt.title('Scatter Plot: Sepal Length vs Petal length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal width')
plt.show
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sl'],df['pw'])
plt.title('Scatter Plot: Sepal Length vs Petal width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal width')
plt.show
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sl'],df['sw'],color='blue')
plt.scatter(df['pl'],df['pw'],color='green')
plt.scatter(df['sl'],df['pl'],color='gray')
plt.scatter(df['sl'],df['pw'],color='yellow')

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sl'],df['sw'],color='blue')
plt.scatter(df['pl'],df['pw'],color='green')
plt.scatter(df['pl'],df['sl'],color='gray')
plt.scatter(df['sw'],df['pw'],color='red')
plt.show()
```

```python
correlation_matrix = df.drop(columns=['class']).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)


plt.title('Correlation Heatmap', fontsize=16)


plt.show()
```

```python
num_df=df.select_dtypes(include=['number'])
plt.figure()
num_df.plot(kind='line')
plt.show()
```

```python
num_df.hist(bins=30,figsize=(8,6))
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x='pl', hue='class', kde=True, element="step")

plt.title('Histogram of Petal Length by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x='pw', hue='class', kde=True, element="step")

plt.title('Histogram of Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Count')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='class', y='sl')
plt.title('Box Plot: Sepal Length Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='class', y='sw')
plt.title('Box Plot: Sepal Width  Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(df, hue='class', diag_kind='kde')
plt.show()
```

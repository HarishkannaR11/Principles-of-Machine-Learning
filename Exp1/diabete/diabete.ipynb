```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('diabete.csv')
df.head()
```

```python
mapping={
    'Yes':1,'No':0,
    'Positive':1,'Negative': 0,
    'Male': 1, 'Female': 0
}
```

```python
df_numeric=df.replace(mapping)
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Polyuria',hue='class')
plt.title('Does Polyuria (Excessive Urination) indicate Diabetes?')
plt.xlabel('Patient has Polyuria?')
plt.ylabel('Count of Patients')
plt.legend(title='Diabetes Result')
plt.show()
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation: Which Symptom Predicts Diabetes Best?')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Histogram of Age, colored by Result
sns.histplot(data=df, x='Age', hue='class', kde=True, element="step", palette='bright')

plt.title('Age Distribution: Positive vs Negative Cases')
plt.xlabel('Age of Patient')
plt.show()
```

```python
# Select only the most important features to keep the plot readable
# 'class' is included to color the dots (hue)
selected_cols = ['Age', 'Polyuria', 'Polydipsia', 'Gender', 'class']

plt.figure(figsize=(10, 8))

# We use the numeric dataframe we created earlier
sns.pairplot(df_numeric[selected_cols], hue='class', diag_kind='kde', palette='husl')

plt.suptitle('Pair Plot of Key Diabetes Factors', y=1.02)
plt.show()
```

```python
sns.lineplot(data=df_numeric, x='Age', y='class', color='crimson')

plt.title('Probability of Diabetes (Positive Class) vs. Age')
plt.ylabel('Probability (0 = Negative, 1 = Positive)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df['Age'])
plt.title('Age Variation Across Patients (Index)')
plt.xlabel('Patient ID (Index)')
plt.ylabel('Age')
plt.show()
```

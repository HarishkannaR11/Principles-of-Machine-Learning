```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('mnist_test.csv')
```

```python
X = df.drop('label', axis=1)
y = df['label']
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X, y)
```

```python
img = X[df['label'] ==3 ].iloc[0].values
plt.imshow(img.reshape(28,28), cmap='gray')
plt.axis('off')
plt.show()
```

```python
X.shape   # (n_samples, 784)
y.unique()
```

```python
import matplotlib.pyplot as plt

y.value_counts().sort_index().plot(kind="bar")
plt.title("Digit Class Distribution")
plt.xlabel("Digit")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(X.values.flatten(), bins=50)
plt.title("Pixel Intensity Distribution")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,4))
for digit in range(10):
    mean_img = X[y == digit].mean().values.reshape(28,28)
    plt.subplot(2,5,digit+1)
    plt.imshow(mean_img, cmap="gray")
    plt.title(f"Digit {digit}")
    plt.axis("off")
plt.show()
```

```python
std_img = X[y == 8].std().values.reshape(28,28)
plt.imshow(std_img, cmap="hot")
plt.title("Pixel Variance for Digit 8")
plt.colorbar()
plt.show()
```
